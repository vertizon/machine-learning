# 概率论

## 概率

### 概率的概念

随机事件的概率的定义为：$E$是随机试验，$S$是样本空间，对于$E$中的每个事件$A$赋予一个实数，记为$P(A)$，称为事件$A$的概率，集合函数$P(*)$满足：
- 非负性
- 规范性
- 可列可加性

可证明，当$n$趋于无穷大时，事件发生的频率在一定意义下接近于概率P$(A)$，基于这一事实，我们有理由将概率$P(A)$用来表征事件$A$在一次试验中发生的可能性的大小（物理意义）。

#### 概念解析

##### 随机现象

- 现象
    - 确定性现象
    - 不确定性现象
        - 随机现象
        
根据是否必然发生，这个世界的现象可以分为两种：确定性现象和不确定现象。确定性现象是：在一定条件下必然发生的现象。反之，是不确定现象。不确定现象中，有一类现象叫随机现象，它满足：**在个别试验中其结果呈现出不确定性，在大量重复试验中其结果有具有统计规律性**。即，这种不确定不是完全的没有规律，而是有一定规律的，而且这些规律是统计规律。

##### 统计规律

定义：大量重复实验或观察中所呈现的固有规律性，就是统计规律性。

这个定义和随机现象基本上构成了循环定义，但可以理解，所谓统计规律，是大量、重复的现象中的规律，这里的关键是大量、重复。

大量：少量则可能不能表现出这个固有的规律。

重复：条件一样才能叫重复。

因此，概率统计提供了描述或量化不确定现象中的规律的能力，不是所有规律，而是大量重复试验中表现出来的规律，其他规律则不能描述。

要找现象中的规律，需要进行试验，而要找到随机现象中的统计规律，则需要的是随机试验，一种满足一定条件的试验。

##### 随机试验


- 可以在相同条件下重复进行
- 每次试验的可能结果不止一个，并且能事先明确试验的所有可能结果
- 进行一次试验之前不能确定哪一个结果会出现

##### 随机事件

随机事件：是样本空间的子集合

**样本空间**：是随机试验所有可能结果的集合

**样本点**：是随机试验的每个结果，记为$e$

随机事件发生：随机事件中的样本点出现

随机事件发生的频率：

事件发生的频数：在相同条件下，进行了n次试验，在n次试验中，事件发生的次数

事件发生的频率：频数/n

频率稳定性：当重复试验的次数逐渐增大时，频率呈现出稳定性，逐渐稳定于某个常数，这种频率稳定性，即通常所说的统计规律性。（这是一种假设，现实现象未必如此，或则说，只有满足这种条件的现象，才是随机现象）

### 相关概念或定理

等可能概型随机试验中的概率。

事件A发生条件下事件B发生的条件概率$P(B|A)=P(AB)/P(A)$

乘法定理：$P(AB) = P(B|A)*P(A)$

全概率公式

贝叶斯公式

两个事件相互独立：$P(AB)=P(A)*P(B)$

## 随机变量

### 随机变量的概念

随机变量是定义在样本空间$S$上的实值单值函数$X(e)$。

#### 概念解析
 
 - 物理意义：将随机试验的结果（即样本点）数量化
 
 - 从定义可知，随机变量的值依赖样本点（试验结果），而试验结果的出现有一定概率，因而随机变量的取值也有一定概率。（这是随机变量区别于一般变量的地方）
 - 单值的意思是，每个样本点只对应一个值
 - 由随机变量表示的事件：随机变量取某个值，对应某些样本点出现，而一个样本点子集合，构成一个随机事件，因此，随机变量取某个值，就对应一个随机事件，一般地（把值扩展为一个集合）：
 
 若$L$是一个实数集合，将$X$在$L$上取值写成$\{X\ \in L \}$，它表示事件$B=\{e|X(e)\in L\}$,即$B$是由$S$中使得$X(e)\in L$的所有样本点$e$所组成的事件，此时有:$P\{X\in L\}=P(B)=P\{e|X(e) \in L \}$
    
- 分类：值是有限个或可列无限多个：离散随机变量，否则：连续随机变量

## 概率分布

概率分布是随机变量的概率分布，描述的是随机变量取一些值的概率。

###  概率质量函数 (probability mass function, PMF)

概率质量函数即分布率，用来描述离散随机变量在不同取值的可能性。定义为：$$P(X=x_k)=p_k,k=1,2,3,...$$ 
$p_k满足1）大于等于0，2）和为1$

例子：
- 0-1分布
- 伯努利分布
- 泊松分布

### 概率密度函数 (probability density function, PDF)

定义为：$$F(x)=\int_{-\infty}^{x} p(t)dt$$

解析：
- 用来描述连续随机变量在不同取值的可能性。
- 并没有直接对特定值给出概率，而是给出了落在区间$dt$内的概率为$p(t)dt$
- $F(x)$是分布函数，定义为：$F(x)=P\{X<=x\}$
- 例子：
    - 均匀分布
    - 指数分布
    - 正态分布
